## Game Mechanics
Brief explanation of game and how it works
#### What is Clash Royale?
Clash Royale is a hybrid card/time strategy game produced by Supercell that is played on mobile devices. The game is played like chess, in the sense that it is 1 on 1, has "pieces"(like a chess piece = knight or queen), and the strategy centers around the interaction of these pieces. You can think about it like rock paper scissors, in the sense that each match is an independent contest between two players.

#### What is the objective of Clash Royale?
Each player has two princess towers and a king tower. The goal is to get the opponent's health on these towers to 0 (or at least lower than their own). The two towers in front of the bigger, central tower are the princess towers. The biggest tower is the king tower. All matches of Clash Royale can range from 30 seconds to 6 minutes total. 
<img src="preview.jpg" style="width:300px;height:400px"/>

#### Rules
- Each player has a budget of "elixir" that steadily increases at 1 elixir/second to a maximum of 10. Each card has an elixir cost, and placing a card on the board will take that cost from your supply.
- Once an opponent's king tower is taken down (regardless of the status of other towers), the opponent automatically wins.
- Each match has 4 stages:
   1. Single elixir (0:00 - 2:00)
       - the rate of elixir increase is 1 elixir/sec
   2. Double elixir (2:00 - 3:00)
       - the rate of elixir increase is 2 elixir/sec
       - Match goes to overtime if no towers are taken down
       - player who has more towers wins (regardless of amount of damage)
   3. Sudden Death (3:00 - 5:00)
       - First player to take a tower wins instantenously
   4. Triple Elixir (5:00 - 6:00)
       - Game is still in sudden death mode
       - rate of elixir increase is 3 elixir/sec

#### What does the player control?
Players also bring with them a deck of 8 cards (that are the equivalent of chess pieces) that each have a special property. There are 103 cards in total that can be split into spells and minions.

```python
def winCounter(data, deck):
    
    #temp variable to store deck we're adding to our barplots
    deck_temp = 0

    #if data is empty
    if data.empty:
        print('There are no battles in the DataFrame!!')
        return 0
    
    #determines who won in the specified deck interaction
    #if person with deck wins, skip
    #because we only care about the cards that make you lose when using that deck
    for i in range(len(data)):
        if data['p1'][i] == deck and data['Winner'][i] == 'p1':
            continue
        else:
            deck_temp = data['p2'][i]
    
    #find a way to convert deck_temp from a string to a list
    deck_temp = ast.literal_eval(deck_temp)
    
    for k in range(len(deck_temp)):
        if deck_temp[k] in spells:
            index = df_spell[df_spell['Spell'] == deck_temp[k]].index
            df_spell['Count'][index]+=1
            
        elif deck_temp[k] in win_condition:
            index = df_WC[df_WC['Win Condition'] == deck_temp[k]].index
            df_WC['Count'][index]+=1
            
        elif deck_temp[k] in troops:
            index = df_troop[df_troop['Troop'] == deck_temp[k]].index
            df_troop['Count'][index]+=1
            
        elif deck_temp[k] in buildings:
            index = df_building[df_building['Building'] == deck_temp[k]].index
            df_building['Count'][index]+=1


    df_spell.sort_values['Count']
    df_WC.sort_values['Count']
    df_troop.sort_values['Count']
    df_building.sort_values['Count']
```
