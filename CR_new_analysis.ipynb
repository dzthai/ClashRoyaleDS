```python
len(df) #2823527
df2 = df.loc[df['average.startingTrophies'] >= 5500] #filter dataframe by skill, indicated by trophy count 5500
len(df2)
```

```python
df2 = df2[['loser.cards.list','winner.cards.list']].copy()
df2 = df2.reset_index(drop=True)
df2
```

```python
df_cards = pd.read_csv('cardmaster.csv')
df_cards = df_cards.rename(columns = {'team.card1.id' : 'id' , 'team.card1.name' : 'name'})
```

```python
df_cards['type'] = "Troop"
df_cards['type'][71:83] = "Building"
df_cards['type'][83:] = "Spell"
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df_cards)
```

```python
#create dictionary to translate (also doing it for troop now)
card_dict = {}
for i in range(len(df_cards)):
    card_dict[df_cards['id'][i]] = df_cards['name'][i]
    
type_dict = {}
for i in range(len(df_cards)):
    type_dict[df_cards['name'][i]] = df_cards['type'][i]
```

```python
type(df2['loser.cards.list'][0]) #string representation of list in each row
example = ast.literal_eval(df2['loser.cards.list'][0]) #convert string rep into actual list
example
```

```python
#translate the list from id to human-readable card names
result = []
for i in example:
    result.append(card_dict[i])
result.sort()
result
```

```python
#combine both previous steps into one function
def translator(row):
    temp = ast.literal_eval(row)
    result = []
    for i in temp:
        result.append(card_dict[i])
    result.sort()
    return result
```

```python
#func to help me debug using apply on each column
def printer(x):
    print(type(x))
    print(x)
```

```python
new_df = df2['loser.cards.list'].apply(translator)
new_df2 = df2['winner.cards.list'].apply(translator)
```

```python
final_df = pd.concat([new_df,new_df2], axis=1)
final_df = final_df.rename(columns = {'loser.cards.list': 'loser', 'winner.cards.list':'winner'},)
final_df
```

```python
#use this to search for games with this deck
deck = ['Skeletons','Cannon','Ice Spirit','Ice Golem','Hog Rider','Fireball','Musketeer','The Log']
deck.sort()

#search for losers who lost using this specific deck
temp_list = list(final_df['loser'])
deck_losers_indices = []

#these are all the games where deck lost
for i in range(len(temp_list)):
    if temp_list[i] == deck:
        deck_losers_indices.append(i)
```

```python
#these are all the games where hog2.6 lost, 'winners' is the decks that won against hog 2.6
winners = []
total_winner_cards = []
for i in deck_losers_indices:
    winners.append(final_df['winner'][i])
    for k in final_df['winner'][i]:
        total_winner_cards.append(k)
```

```python
counter = dict.fromkeys(df_cards['name'], 0)
for i in total_winner_cards:
    counter[i] += 1
{k: v for k, v in sorted(counter.items(), key=lambda item: item[1])}
```

```python
#side project: seeing which cards are way too strong in the meta by counting occurences 
#of cards from the winner list

final_df['winner']
strong_meta = []
for i in range(len(final_df['winner'])):
    for k in final_df['winner'][i]:
        strong_meta.append(k)

        
meta_counter = dict.fromkeys(df_cards['name'], 0)
for i in strong_meta:
    meta_counter[i] += 1
#{k: v for k, v in sorted(meta_counter.items(), key=lambda item: item[1])}
```

```python
#deriving win rate % to check which cards are performing well in the meta and which cards need balancing
win_rates = {}
for key, val in meta_counter.items():
    win_rates[key] = (val / 370033) * 100 #this is the total number of games from the previous dataframe
    
win_rates
```
